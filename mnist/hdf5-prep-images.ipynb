```julia
using Images, FileIO, HDF5
```

```julia
n_data = 2
```

```julia
cat_img = load("cat.jpg")
```

```julia
Gray.(cat_img)
```

```julia
cat_img = convert(Array{Float32,2}, Gray.(cat_img))
```

```julia
cat_img2 = load("cat1.jpg")
```

```julia
Gray.(cat_img2)
```

```julia
cat_img2 = convert(Array{Float32,2}, Gray.(cat_img2))
```

```julia
h5open("data/cat.hdf5", "w") do h5
    dset_data = d_create(h5, "data", datatype(Float32), dataspace(400, 600, 1, n_data))
    dset_label = d_create(h5, "label", datatype(Float32), dataspace(1, n_data))
end
```

```julia
catfile = h5open("data/cat.hdf5", "r+")
```

```julia
catfile["data"]
```

```julia
catfile["data"][:,:,1,1] = cat_img
```

```julia
catfile["data"][:,:,1,2] = cat_img2
```

```julia
close(catfile)
```
